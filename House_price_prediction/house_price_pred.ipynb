```python
drop_features = []
for column in df.columns:
    if df[column].isnull().sum()>1000:
        drop_features.append(column)
drop_features
```

```python
df[drop_features].sample(5)
```

```python
df.drop(columns=drop_features,axis=1,inplace=True)
```

```python
df.shape
```

```python
df.head(5)
```

```python
categorical_features = []
numerical_features = []

for features in df.columns:
    if df[features].dtype == 'O':
        categorical_features.append(features)
    else:
        numerical_features.append(features)
```

```python
df[categorical_features].isnull().sum()
```

```python
missing_categorical = []

for feature in categorical_features:
    if df[feature].isnull().sum()>0:
        missing_categorical.append(feature)

for feature in missing_categorical:
    print("number of rows :",df[feature].count())
    print(df[feature].value_counts())
```

```python
mode_entry = ['BsmtQual','BsmtFinType1','BsmtFinType2','GarageFinish','GarageQual']

df['BsmtQual'].fillna('TA',inplace=True)
missing_categorical.remove('BsmtQual')
df['BsmtFinType1'].fillna('Unf',inplace=True)
missing_categorical.remove('BsmtFinType1')
df['BsmtFinType2'].fillna('Unf',inplace=True)
missing_categorical.remove('BsmtFinType2')
df['GarageFinish'].fillna('SBrkr',inplace=True)
missing_categorical.remove('GarageFinish')
df['GarageQual'].fillna('TA',inplace=True)
missing_categorical.remove('GarageQual')
```

```python
for feature in missing_categorical:
    df[feature].fillna("Missing_"+feature,inplace=True)
```

```python
df[categorical_features].isnull().sum()
```

```python
missing_numerical = []

for feature in numerical_features:
    if df[feature].isnull().sum()>0:
        missing_numerical.append(feature)

missing_numerical
```

```python
df[missing_numerical].isnull().sum()
```

```python
df[missing_numerical].describe()
```

```python
for feature in missing_numerical:
    plt.scatter(range(len(df[feature])),df[feature])
    plt.title(feature)
    plt.show()
```

```python
for feature in missing_numerical:
    df[feature].fillna(df[feature].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Id'],axis=1,inplace=True)
```
